####**Análisis de datos**

##### Ingreso total por cada tienda

```python
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

ingresos_totales = pd.Series({
    'Tienda 1': ingreso_tienda1,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
})

for tiendas, ingresos_totales in ingresos_totales.items():
    print(f"Ingreso {tiendas} : ${ingresos_totales:,.0f}.-")
```

##### Tiendas Con Mayor y Menor Ingresos

```python
max_income_store = ingresos_totales.idxmax()
max_income_value = ingresos_totales.max()

min_income_store = ingresos_totales.idxmin()
min_income_value = ingresos_totales.min()

print(f"Tienda Mayor Ingreso {max_income_store} Con total de ${max_income_value:,.0f}.-")
print(f"Tienda Menor Ingreso {min_income_store} Con total de ${min_income_value:,.0f}.-")
```

#####Gráfico de Barra de los ingresos por tienda

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
x = range(len(tiendas))
width = 0.35
bar_colors = ['tab:blue', 'tab:green', 'tab:orange', 'tab:red']

plt.figure(figsize=(12, 8))

bars = plt.bar(x, ingresos_totales.values, color=bar_colors)
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total')
plt.title('Ingresos Totales por Tienda')
plt.bar_label(bars, fmt='{:,.0f}')
plt.xticks(x, tiendas)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio = pd.DataFrame(columns=['Tienda', 'Promedio'])
def promedio_calificacion(dataset):
  return dataset['Calificación'].mean()

promedio_t1 = promedio_calificacion(tienda1)
promedio_t2 = promedio_calificacion(tienda2)
promedio_t3 = promedio_calificacion(tienda3)
promedio_t4 = promedio_calificacion(tienda4)


promedio = pd.Series({
    'Tienda 1': promedio_t1,
    'Tienda 2': promedio_t2,
    'Tienda 3': promedio_t3,
    'Tienda 4': promedio_t4
})

for tiendas, promedio in promedio.items():
    print(f"Calificación Promedio {tiendas} : {promedio:,.5f} %")
```

#####Gráfico Promedio por Tienda

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.figure(figsize=(8,4))
bars = plt.barh(tiendas, promedio.values , color= mcolors.BASE_COLORS) # Use plt.barh for horizontal bars
plt.xlabel('Calificación promedio') # X-axis label for horizontal bar chart
plt.ylabel('Tiendas') # Y-axis label for horizontal bar chart
plt.title('Calificación Promedio de los clientes para cada tienda')

# Correctly position labels for horizontal bars
for bar in bars:
    width = bar.get_width()
    plt.text(width , bar.get_y() + bar.get_height()/2, f'{width:.3f}',rotation=45, va='center', ha='left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def tienda_top(dataset):
  return dataset['Producto'].value_counts().head(1)
def tienda_low(dataset):
  return dataset['Producto'].value_counts().tail(1)
top_tienda1 = tienda_top(tienda1)
low_tienda1 = tienda_low(tienda1)
top_tienda2 = tienda_top(tienda2)
low_tienda2 = tienda_low(tienda2)
top_tienda3 = tienda_top(tienda3)
low_tienda3 = tienda_low(tienda3)
top_tienda4 = tienda_top(tienda4)
low_tienda4 = tienda_low(tienda4)


print(f"Tienda 1:")
print(f"MÁS vendido  : {top_tienda1.index[0]} ({top_tienda1.values[0]} ventas)")
print(f"MENOS vendido: {low_tienda1.index[0]} ({low_tienda1.values[0]} ventas)\n")

print(f"Tienda 2:")
print(f"MÁS vendido  : {top_tienda2.index[0]} ({top_tienda2.values[0]} ventas)")
print(f"MENOS vendido: {low_tienda2.index[0]} ({low_tienda2.values[0]} ventas)\n")

print(f"Tienda 3:")
print(f"MÁS vendido  : {top_tienda3.index[0]} ({top_tienda3.values[0]} ventas)")
print(f"MENOS vendido: {low_tienda3.index[0]} ({low_tienda3.values[0]} ventas)\n")

print(f"Tienda 4:")
print(f"MÁS vendido  : {top_tienda4.index[0]} ({top_tienda4.values[0]} ventas)")
print(f"MENOS vendido: {low_tienda4.index[0]} ({low_tienda4.values[0]} ventas)\n")
```
